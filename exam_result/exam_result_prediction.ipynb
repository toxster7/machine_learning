# Predictions of exam results based on study hours and previous exam scores using K-nearest neighbour classification model 

```python
#Import libriries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Download data
df = pd.read_csv('student_exam_data.csv')
df.head()
```

```python
#Checking for NaN values in dataframe
df.isnull().sum()
```

```python
X=df.iloc[:,:-1]#independet variable
y=df.iloc[:,2]#dependet(target)
```

```python
from sklearn.model_selection import train_test_split
#Splitting data to train and test dataframes
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Tuning k-estimator for KNN Model
train_score = {}
test_score = {}
n_neighbors = np.arange(2, 30, 1) #array of int from 2 to 30
for neighbor in n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=neighbor)
    knn.fit(X_train, y_train)
    train_score[neighbor]=knn.score(X_train, y_train)
    test_score[neighbor]=knn.score(X_test, y_test)
```

```python
#Plotting accuracys scores for each k-estimator
plt.plot(n_neighbors, train_score.values(), label="Train Accuracy")
plt.plot(n_neighbors, test_score.values(), label="Test Accuracy")
plt.xlabel("Number Of Neighbors")
plt.ylabel("Accuracy")
plt.title("KNN: Varying number of Neighbors")
plt.legend()
plt.xlim(0, 33)
#plt.ylim(0.8, 1)
plt.grid()
plt.show()
```

```python
#Find the best k-estimator
for key, value in test_score.items():
    if value==max(test_score.values()):
        print(key)
```

```python
from sklearn.model_selection import GridSearchCV,KFold

#Tuning k-estimator for KNN Model using sklearn.GridSearch

kf=KFold(n_splits=5,shuffle=True,random_state=42)
parameter={'n_neighbors': np.arange(2, 30, 1)}
knn=KNeighborsClassifier()
knn_cv=GridSearchCV(knn, param_grid=parameter, cv=kf, verbose=1)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_params_)
```

```python
from mlxtend.plotting import plot_decision_regions

#Training the model with tuned k-estimator from GridSearch
xx = X_train.values
yy = y_train.astype(int).values
clf = KNeighborsClassifier(n_neighbors=4)
clf.fit(xx, yy)
# Plotting decision region for train data
plot_decision_regions(xx, yy, clf=clf, legend=2)
# Adding axes annotations
plt.xlabel('Study Hours')
plt.ylabel('Previous Exam Score')
plt.title('Train-Knn with K=' + str(3))
plt.show()
```

```python
#Plotting test datar
plot_decision_regions(X_test.values, y_test.values, clf=clf, legend=2)
# Adding axes annotations
plt.xlabel('Study Hours')
plt.ylabel('Previous Exam Score')
plt.title('Test-Knn with K=' + str(4))
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#Making predictions on the test data
y_pred = clf.predict(X_test)

#Building confusion matrix of predictions
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import (precision_score,recall_score,f1_score)

#Evaluating model using Precision,Recall,F1

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1score = f1_score(y_test, y_pred)
 
print(f"Precision = {precision}")
print(f"Recall = {recall}")
print(f"F1 Score = {f1score}")
```
